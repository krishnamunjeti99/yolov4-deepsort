```python
def get_image_from_video(path,save_path):
    capture = cv2.VideoCapture(path)
    os.makedirs(save_path,exist_ok=True)
    count=59
    file_count=1
    while(capture.isOpened()):
        ret, frame = capture.read()
        count+=1
        if count==60:
            if ret==True:
                cv2.imwrite(os.path.join(save_path,"polyp_"+str(file_count)+".png"),frame)
                print("polyp_"+str(file_count)+".png")
                file_count+=1
            else:
                break
            count=0

    capture.release()
```

```python
import cv2
import numpy as np
import os
import sys
def get_image_from_video_all(path,save_path,name):
    capture = cv2.VideoCapture(path)
    os.makedirs(save_path,exist_ok=True)
    file_count=1
    while(capture.isOpened()):
        ret, frame = capture.read()
        if ret==True:
            cv2.imwrite(os.path.join(save_path,name+"_"+str(file_count)+".png"),frame)
            #print("polyp_"+str(file_count)+".png")
            file_count+=1
        else:
            break
    capture.release()
    print(name + " finished")
```

```python
import os
path = "outputs"
for i in os.listdir(path):
    get_image_from_video_all(os.path.join(path,i),os.path.join("outputs_capture",i[:-4]),i[:-4])
```
